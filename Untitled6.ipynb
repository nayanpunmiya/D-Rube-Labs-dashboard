```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn import tree
from sklearn.feature_selection import SelectFromModel
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from sklearn.ensemble import ExtraTreesClassifier, AdaBoostClassifier, GradientBoostingClassifier, RandomForestClassifier
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN

# Load the dataset
malds = pd.read_csv("/content/MalwareData.csv", sep="|", low_memory=True)

# Clean the dataset and select top features
X = malds.drop(['Name', 'md5', 'legitimate'], axis=1).values
y = malds['legitimate'].values
extratrees = ExtraTreesClassifier().fit(X, y)
model = SelectFromModel(extratrees, prefit=True)
X_new = model.transform(X)
newfeatures = X_new.shape[1]

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2, random_state=42)

# Scale the data
scaler = preprocessing.StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

# Perform PCA to reduce the data to 2 dimensions for visualization
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# Perform PCA for explained variance analysis
pca_full = PCA()
pca_full.fit(X_train)

# Plot PCA Explained Variance and Cumulative Explained Variance
def plot_pca_variance(pca_model):
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(np.arange(1, len(pca_model.explained_variance_ratio_) + 1), pca_model.explained_variance_ratio_, 'o-')
    plt.title('PCA Explained Variance')
    plt.xlabel('Number of Components')
    plt.ylabel('Explained Variance Ratio')

    plt.subplot(1, 2, 2)
    plt.plot(np.arange(1, len(pca_model.explained_variance_ratio_) + 1), np.cumsum(pca_model.explained_variance_ratio_), 'o-')
    plt.axhline(y=0.90, color='r', linestyle='--', label='90% Explained Variance')
    plt.title('PCA Cumulative Explained Variance')
    plt.xlabel('Number of Components')
    plt.ylabel('Cumulative Explained Variance')
    plt.legend(loc='lower right')
    plt.tight_layout()
    plt.show()

# Call function to plot PCA variance
plot_pca_variance(pca_full)

# Print PCA components
print("Explained Variance Ratio of PCA components:", pca.explained_variance_ratio_)
print("PCA 1 and PCA 2 values (first 5 samples):")
print(X_train_pca[:5])

# Function to plot confusion matrix
def plot_confusion_matrix(cm, algo_name):
    plt.figure(figsize=(6, 5))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title(f'Confusion Matrix - {algo_name}')
    plt.colorbar()
    tick_marks = np.arange(len(np.unique(y)))
    plt.xticks(tick_marks, np.unique(y), rotation=45)
    plt.yticks(tick_marks, np.unique(y))

    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, f"{cm[i, j]} ({cm_normalized[i, j]:.2f})",
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
    plt.show()

# Function to plot SVM Hyperplane
def plot_svm_hyperplane(X, y, model):
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X)
    model.fit(X_pca, y)
    x_min, x_max = X_pca[:, 0].min() - 1, X_pca[:, 0].max() + 1
    y_min, y_max = X_pca[:, 1].min() - 1, X_pca[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.8)
    plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, edgecolors='k', marker='o', s=30)
    plt.title(f"SVM Decision Boundary and Hyperplane")
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.show()

# Classification models
models = {
    "DecisionTree": tree.DecisionTreeClassifier(max_depth=10),
    "RandomForest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Adaboost": AdaBoostClassifier(n_estimators=50),
    "GradientBoosting": GradientBoostingClassifier(n_estimators=50),
    "LogisticRegression": LogisticRegression(max_iter=1000),
    "SVM": SVC(kernel='linear', C=1.0, tol=1e-2, random_state=42),
    "GNB": GaussianNB(),
    "KNN": KNeighborsClassifier(n_neighbors=5)
}

results = {}
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Main loop for classification models
for algo, clf in models.items():
    clf.fit(X_train_pca, y_train)
    y_pred_binary = clf.predict(X_test_pca)

    accuracy = accuracy_score(y_test, y_pred_binary)
    precision = precision_score(y_test, y_pred_binary)
    recall = recall_score(y_test, y_pred_binary)
    f1 = f1_score(y_test, y_pred_binary)
    conf_matrix = confusion_matrix(y_test, y_pred_binary)

    # Store results for comparison
    results[algo] = {
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1
    }

    print(f"{algo} - Accuracy: {accuracy:.4f}")
    print(f"{algo} - Precision: {precision:.4f}")
    print(f"{algo} - Recall: {recall:.4f}")
    print(f"{algo} - F1 Score: {f1:.4f}")
    print(f"{algo} - Confusion Matrix:\n{conf_matrix}\n")

    # Plot Confusion Matrix (for algorithms other than DecisionTree and RandomForest)
    if algo not in ["DecisionTree", "RandomForest"]:
        plot_confusion_matrix(conf_matrix, algo)

    # Plot SVM Hyperplane for SVM only
    if algo == "SVM":
        plot_svm_hyperplane(X_train, y_train, clf)

    # Visualize a few decision trees in the forest with improved readability for Random Forest
    if algo == "RandomForest":
        for i in range(3):
            plt.figure(figsize=(18, 14), dpi=120)
            tree.plot_tree(clf.estimators_[i], filled=True, max_depth=3, feature_names=[f'Feature {i}' for i in range(X_new.shape[1])], class_names=['Legitimate', 'Malware'], fontsize=10)
            plt.title(f"Tree {i+1} in Random Forest (Max Depth=3)")
            plt.show()

    # Plot Decision Tree with high readability for DecisionTree model
    if algo == "DecisionTree":
        plt.figure(figsize=(18, 12), dpi=120)
        tree.plot_tree(clf, filled=True, feature_names=[f'Feature {i}' for i in range(X_new.shape[1])], class_names=['Legitimate', 'Malware'], max_depth=5, fontsize=10)
        plt.title("Decision Tree Visualization (Max Depth=5)")
        plt.show()

# Comparison of all algorithms
# Display comparison in a DataFrame and plot the metrics

comparison_df = pd.DataFrame(results).T
print("\nComparison of All Algorithms:\n", comparison_df)

# Plot comparison metrics for all models
comparison_df.plot(kind='bar', figsize=(12, 8))
plt.title("Comparison of Classification Models")
plt.xlabel("Algorithm")
plt.ylabel("Score")
plt.legend(loc="upper right")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
